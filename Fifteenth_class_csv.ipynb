```python
#list inside dict
import pandas as pd
data = {
    'sn': [1,2,3,4,5],
    'name': ['Ram', 'Shyam', 'Sita', 'Hari', 'Gita'],
    'add': ['Bhaktapur', 'Lalipur', 'Jhapa', 'Chitwan', 'Kathmandu'],
    'age': [32, 34, 47, 38, 29]
}
df = pd.DataFrame(data)
df.to_csv('info.csv')
df
```

```python
#list inside list
data = [[1, 'Ram', 34, 'Kathmandu'],
        [2, 'Shyam', 43, 'Bhaktapur'],
        [3, 'Sita', 39, 'Chitwam'],
       [4, 'Gita', 28, 'Lalitpur']]
col = ['sn', 'name', 'age', 'add']
df = pd.DataFrame(data, columns=col)
df.to_csv('info1.csv')
df
```

```python
import csv
csv_columns = ['Name', 'Age', 'Address']
dict = [
    {'Name':'Aayush','Age':23,'Address':'Katmandu'},
    {'Name':'Shyam','Age':33,'Address':'Patan'}
]

csvfile = open("data.csv", 'w')
writer = csv.DictWriter(csvfile, fieldnames=csv_columns)
writer.writeheader()
for data in dict:
    writer.writerow(data)
csvfile.close()
```

```python
# concat

df1 = pd.DataFrame({'sn': [1,2,3],
                    'name': ['Ram', 'Shyam', 'Hari'],
                    'add': ['Kathmandu', 'Bhaktapur', 'Lalitpur']})

df2 = pd.DataFrame({'sn': [4,5,6],
                    'name': ['Sita', 'Gita', 'Krishna'],
                    'add': ['Chitwan', 'Jhapa', 'Banepa']})

df = pd.concat([df1, df2], axis = 0, ignore_index=True)
df
```
